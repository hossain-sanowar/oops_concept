```python
class TensorShowcase:
    def __init__(self):
        """Initialize the class with various tensor ranks."""
        # Rank 0: scalar
        self.rank_0 = tf.constant(4)

        # Rank 1: Vector
        self.rank_1 = tf.constant([1,2,3])

        # Rank 2: Matrix (3x3)
        self.rank_2 = tf.constant([[4,5,6], [7,8,9]], dtype=tf.int32)

        # Rank 3: 3D Tensor (2x2x2)
        self.rank_3 = tf.constant([
                                [[1,2], [3,4]],
                                [[5,6], [7,8]]
                                ], dtype=tf.int32)



    def display_shapes(self):
        """Method to print the shapes of all stored tensors."""
        print(f"Rank 0 Shape: {self.rank_0.shape} (Scalar)")
        print(f"Rank 1 Shape: {self.rank_1.shape} (Vector)")
        print(f"Rank 2 Shape: {self.rank_2.shape} (Tensor)")
        print(f"Rank 3 Shape: {self.rank_3.shape} (Tensor)")

    def create_tensor(self, rows, cols):
        shape=(rows, cols)

        random_tensor = tf.random.normal(shape=shape, mean=0, stddev=1)
        return random_tensor


    def zeros_create(self, rows, cols):
        shape=(rows, cols)
        zeros_tensor = tf.zeros(shape=shape, dtype=tf.float32)
        return zeros_tensor

    def ones_create(self, rows, cols):
        shape=(rows, cols)
        ones_tensor = tf.ones(shape=shape, dtype=tf.float32)
        return ones_tensor

    # Random uniform distribution between 0 and 10
    def create_uniform(self, rows, cols):
        shape=(rows, cols)
        uniform_tensor = tf.random.uniform(shape=shape, minval=0, maxval=1)
        return uniform_tensor


tensor=TensorShowcase()
tensor.display_shapes()
tensor.create_tensor(5,3)
tensor.zeros_create(5,3)
tensor.ones_create(5,3)
tensor.create_uniform(5,3)
```
