```python
#define a class
class model(object):
    # constructor method
    def __init__(self, numlayers, numunits, name):
        self.numlayers = numlayers
        self.numunits = numunits
        self.name = name
        self.weights = 10

    def howManyUnits(self):
        totalUnits=self.numlayers*self.numunits
        print(f'There are a total of {totalUnits} units in this model')

    def trainTheModel(self, x):
        self.weights += x
        return self.weights

    def __str__(self):
        return f'This is a {self.name} model.'
```

```python
# create an instance of the class
ex=model(8,17,'LLM')
ex.howManyUnits()
```

```python
str(ex)
```

```python
#ex.weights()
#print(ex.weights)
ex.trainTheModel(2)
print(ex.weights)
```

```python
class Calculator(object):
    def __init__(self,num1,num2):
        self.num1 = num1
        self.num2 = num2
        self.sum = 0

    def add(self):
        return self.num1 + self.num2

    def sub(self):
        return self.num1 - self.num2
    def mul(self):
        return self.num1 * self.num2
```

```python
cal=Calculator(2,3)

print(f"add: {cal.add()}")
print(f"substraction: {cal.sub()}")
print(f"multiplication: {cal.mul()}")
```
