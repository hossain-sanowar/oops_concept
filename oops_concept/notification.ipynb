```python
from abc import ABC, abstractmethod
class Notification(ABC):
    def __str__(self,recipient, message):
        self.recipient = recipient
        self.message = message

    @abstractmethod
    def send(self):
        pass

    def log(self):
        """Concrete method - available to all subclasses"""
        return f"Notification logged for {self.recipient}"

print("✓ Abstract class created successfully!")
print()

# TEST PIECE 1: Can we instantiate abstract class?
try:
    notification = Notification("test", "message")
    print("✗ ERROR: Should not be able to create abstract Notification!")
except TypeError as e:
    print(f"✓ Good! Cannot instantiate abstract class: {e}")
print()
```

```python
# ============================================
# PIECE 2: Create Email Notification
# ============================================
print("PIECE 2: Testing Email Notification...")

class EmailNotification(Notification):
    """Concrete implementation for email notifications"""

    def __init__(self, recipient, message, subject):
        super().__init__(recipient, message)
        self.subject = subject

```
