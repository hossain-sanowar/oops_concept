# Key Concepts:

- Abstract Base Class (ABC) - Use ABC from the abc module as a parent class
- @abstractmethod - Decorator that marks methods that must be implemented by subclasses
- Cannot instantiate - You can't create objects directly from abstract classes
- Enforces structure - Ensures all subclasses implement required methods

```python
from abc import ABC, abstractmethod
```

```python
class Shape(ABC):
    """Abstract base class for shapes"""

    @abstractmethod
    def area(self):
        """Calculate area -must be implemented by subclass"""
        pass

    @abstractmethod
    def perimeter(self):
        """Calculate perimeter -must be implemented by subclass"""
        pass

    @abstractmethod
    def describe(self):
        """Concrete method - can be used by all classes"""
        pass

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)

    def describe(self):
        return f'Rectangle: ({self.width}, {self.height})'



```

```python
rect=Rectangle(20,20)
print(rect.area())
print(rect.perimeter())
print(rect.describe())
```

```python
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return self.radius * self.radius

    def area_circle(self):
        return 3.14 * self.radius * self.radius

    def perimeter(self):
        return 2 * self.radius * self.radius

    def describe(self):
        return f'Circle: ({self.radius}, {self.radius})'

    def square(self):
        return self.radius * self.radius
```

```python
circ=Circle(15)
print(circ.area())
print(circ.perimeter())
print(circ.describe())
print(circ.area_circle())
print(circ.square())
```

```python
# Example 2: Abstract Class with Properties
class Vehicle(ABC):
    """Abstract vehicle class"""
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model


    @abstractmethod
    def start_engine(self):
        pass
    @abstractmethod
    def stop_engine(self):
        pass

    @property
    @abstractmethod
    def fuel_type(self):
        """Abstract property"""
        pass

class Car(Vehicle):
    def __init__(self, brand, model, fuel):
        super().__init__(brand, model)
        self._fuel_type = fuel

    def start_engine(self):
        return f"{self.brand} {self.model} engine starting"
    def stop_engine(self):
        return f"{self.brand} {self.model} engine stopping"

    @property
    def fuel_type(self):
        return self._fuel_type





```

```python
car=Car("BMW", "X1", "benzin")
print(car.start_engine())
print(car.stop_engine())
print(car.fuel_type)
```

```python
# Example 3: Payment Processing System
class PaymentProcessor(ABC):
    """Abstract payment processor class"""
    @abstractmethod
    def validate_payment(self, amount):
        pass
    @abstractmethod
    def process_payment(self, amount):
        pass
    @abstractmethod
    def refund_payment(self, transaction_id):
        pass

class CreditCardPaymentProcessor(PaymentProcessor):
    def validate_payment(self, amount):
        if amount < 0:
            return False
        return True
    def process_payment(self, amount):
        if self.validate_payment(amount):
            return f"Processing payment of {amount} credit card"
        return f"Invalid payment of {amount} credit card"
    def refund_payment(self, transaction_id):
        return f"Refund payment of {transaction_id} credit card"

class PayPalProcessor(PaymentProcessor):
    def validate_payment(self, amount):
        if amount < 0:
            return False
        return True
    def process_payment(self, amount):
        if self.validate_payment(amount):
            return f"Processing payment of {amount} via PayPal"
        return f"Invalid payment of {amount} via PayPal"
    def refund_payment(self, transaction_id):
        return f"Refund payment of {transaction_id} via PayPal"
```

```python
card_payment_processor=CreditCardPaymentProcessor()
print(card_payment_processor.validate_payment(-100))
print(card_payment_processor.process_payment(100))
print(card_payment_processor.refund_payment(-100))
```
